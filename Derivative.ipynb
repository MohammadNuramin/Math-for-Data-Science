## Dervatives

```python
x = np.arange(0,7,.1)
y = 2*x
```

```python
def inte():  
    slice_area_all = np.zeros(y.shape[0])
    for i in range (1,len(x)):
        slice_area_all[i] = 1*y[i-1]
        
    area = slice_area_all.cumsum()
    
    return area
```

```python
area = inte()
print(area)
```

```python
plt.plot(x, x**2, label='True')
plt.plot(x, area, label='Estimated')
plt.show()
```
